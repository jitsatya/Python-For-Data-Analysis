```python
import numpy as np
import pandas as pd

deliveries = pd.read_csv('Datasets/deliveries.csv')
deliveries.head()

matches = pd.read_csv('Datasets/matches.csv')
matches.head()
```

```python
# Merge
new = deliveries.merge(matches, left_on='match_id', right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum() # multi_index series
new1 = new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending = False).reset_index() # convert to dataframe

new2 = new1.drop_duplicates(subset='season', keep='first').sort_values('season')
new2[['season', 'batsman']]
```
