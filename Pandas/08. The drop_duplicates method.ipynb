```python
import numpy as np
import pandas as pd

data = pd.read_csv("Datasets/matches - matches.csv")
data.head(3)
```

```python
# drop duplicates on a column
data.drop_duplicates(subset=['city']).shape[0]
```

```python
# drop duplicates on multiple columns
# It is like and operation, city and season together are dulpicates

data.drop_duplicates(subset=['city', 'season']).head(5)
```

```python
#to get the winner of all the IPL seasons

data.drop_duplicates(subset=['season'], keep='last')[['season', 'winner']].sort_values(['season'])
```
