```python
import numpy as np
import pandas as pd
data = pd.read_csv('Datasets/deliveries.csv')
data.head()
```

```python
# Find the most destructive death over batsman in the history of IPL
# Strike Rate- Number of runs/ number of balls/100
# Min 200 balls  in over 16-20

death_over = data[data['over']>15]

all_batsman = death_over.groupby('batsman')['ball'].count()

x = all_batsman>200

batsman_list = all_batsman[x].index.tolist()
```

```python
# Runs scored by all in batsman_list
# Balls played by the batsmans

final = data[data['batsman'].isin(batsman_list)]

runs = final.groupby('batsman')['batsman_runs'].sum()

final = final[final['wide_runs']==0]
final = final[final['bye_runs']==0]
final = final[final['legbye_runs']==0]
final = final[final['penalty_runs']==0]

balls = final.groupby('batsman')['ball'].count()

strike_rate = (runs/balls)*100
strike_rate.sort_values(ascending=False).head()

```
