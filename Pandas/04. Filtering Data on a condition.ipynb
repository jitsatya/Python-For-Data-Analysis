```python
import numpy as np
import pandas as pd

data = pd.read_csv("Datasets/matches - matches.csv")
```

```python
#masking
mask = data['city']=='Hyderabad' # Boolean series
data[mask].head(3)
```

```python
# Function to find number of matches in each city

def get_city_match_count(city):
    mask = data['city'] == city
    return data[mask].shape[0]

print(get_city_match_count('Pune'))
print(get_city_match_count('Mumbai'))
print(get_city_match_count('Kolkata'))
print(get_city_match_count('Bangalore'))
```

```python
#Multiple conditions

mask1 = data['city'] == 'Mumbai'
mask2 = data['date'] > '2010-01-01'
data[mask1 & mask2].head(3)
```

```python
def get_match_count_city_year(city, date):
    mask1 = data['city'] == city
    mask2 = data['date'] > date
    return data[mask1 & mask2].shape[0]

print(get_match_count_city_year('Mumbai', '2015-01-01'))
print(get_match_count_city_year('Bangalore', '2015-01-01'))
print(get_match_count_city_year('Chennai', '2015-01-01'))
```
