```python
import numpy as np
import pandas as pd

data = pd.read_csv("Datasets/matches - matches.csv")
data.head(3)
```

```python
data['team1'].value_counts() + data['team2'].value_counts()
```

```python
# finding top 5 teams
(data['team1'].value_counts() + data['team2'].value_counts()).head(5) # not sorted hence result is wrong
```

```python
# finding top 5 teams

(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False) # sorting descending
```

```python
data.sort_values('city', ascending=False).head(3)

# This change is not parmanent
```

```python
# to make the change parmanent while being ultra careful
data.sort_values('city', inplace = True)
data.head(4) # data is parmanently changed in the dataframe
```

```python
#Sort on multiple columns
data.sort_values(['city', 'date']).head(4) # sorted by city and then date
```

```python
# Handling order of sort on multiple columns
data.sort_values(['city', 'date'], ascending=[True, False]).head(5)
```
