### Plotting Graphs using NumPy and MatPlotlib

```python
import numpy as np
```

```python
x = np.linspace(-40,40,100) # Creating array from -40 to 40 with 100 elements in it and linspaced
print(x)
print(x.size)
```

```python
y = np.sin(x) # Creating y with sin(x)
print(y.size)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline #for displaying graph in screen
```

```python
plt.plot(x,y) # Plotting graph
```

```python
y = x*x+2*x+6 #Quadratic equation formula
```

```python
plt.plot(x,y) #Graph of quadratic equation Parabola
```
